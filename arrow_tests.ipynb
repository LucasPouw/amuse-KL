```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2*np.pi,10*np.pi,100)
y = np.sin(x)

fig,ax = plt.subplots()
ax.plot(x,y)
inv = ax.transData.inverted()
test = inv.transform([1,1])
print(test)
ax.text(test[0],test[1],'test')
plt.show()
```

```python
x = np.arange(0, 10, 0.005)
y = np.exp(-x/2.) * np.sin(2*np.pi*x)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
ax.set_xlim(0, 10)
ax.set_ylim(-1, 1)

# print( ax.transLimits.transform([(5, 0), (1,2)])) #takes you from data to axes coordinates
inv = ax.transLimits.inverted() #takes you from axes to data coordinates
# print(inv.transform((0.5,0.5)))
# print(inv.transform((0,0)))

arrowbase = np.array((0.5,0.5)) #in axis coordinates
arrowhead = np.array((0.4,0.4)) #in axis coordinates
ax.arrow(*inv.transform(arrowbase),*inv.transform(arrowhead))
print(inv.transform(arrowhead))
print(inv.transform(arrowhead) - inv.transform(arrowbase))
ax.scatter(*inv.transform(arrowbase),marker='x',color='black')
ax.scatter(*inv.transform(arrowhead),marker='^',color='black')

ax.annotate("", xy=arrowhead, xytext=arrowbase,
            arrowprops=dict(arrowstyle="->"),xycoords='axes fraction')

plt.show()
```
